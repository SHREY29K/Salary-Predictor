```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv("survey_results_public.csv")
```

```python
dataset.head()
```

```python
dataset = dataset[["Country", "EdLevel", "YearsCodePro", "Employment", "ConvertedCompYearly"]]
dataset = dataset.rename({"ConvertedCompYearly": "Salary"}, axis=1)
dataset.head()
```

```python
dataset = dataset[dataset["Salary"].notnull()]
dataset.head()
```

```python
dataset.info()
```

```python
dataset = dataset.dropna()
dataset.isnull().sum()
```

```python
dataset = dataset[dataset["Employment"] == "Employed, full-time"]
dataset = dataset.drop("Employment", axis=1)
dataset.info()
```

```python
dataset['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categorical_map={}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categorical_map[categories.index[i]] = categories.index[i]
        else:
            categorical_map[categories.index[i]] = 'Other'
    return categorical_map
```

```python
country_map = shorten_categories(dataset.Country.value_counts(), 400)
dataset['Country'] = dataset['Country'].map(country_map)
dataset.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 18))
dataset.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=-90)
plt.show()
```

```python
# Filter Salary values within the range of 10,000 to 25,000
dataset = dataset[(dataset["Salary"] <= 25000) & (dataset["Salary"] >= 10000)]

# Remove rows where the Country is "Other"
dataset = dataset[dataset['Country'] != 'Other']
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 18))
dataset.boxplot('Salary', 'Country', ax=ax)
plt.suptitle('Salary (US$) v Country')
plt.title('')
plt.ylabel('Salary')
plt.xticks(rotation=-90)
plt.show()
```

```python
dataset["YearsCodePro"].unique()
```

```python
def clean_experience(x):
    if x=="More than 50 years":
        return 50
    if x=="Less than 1 year":
        return 0.5
    return float(x)

dataset['YearsCodePro'] = dataset['YearsCodePro'].apply(clean_experience)
```

```python
dataset["EdLevel"].unique()
```

```python
def clean_education(x):
    # Normalize the string
    x = x.lower().strip()  # Convert to lowercase and strip whitespace
    x = x.replace("’", "'")  # Replace smart quotes with standard quotes

    # Check for various education levels using keywords
    if "bachelor" in x:
        return "Bachelor's degree"
    elif "master" in x or "m.a." in x or "m.s." in x or "m.eng." in x or "mba" in x:
        return "Master's degree"
    elif "professional" in x or "jd" in x or "md" in x or "ph.d" in x or "ed.d" in x:
        return "Post Grad"
    
    return "Less than Bachelor's degree"

# Apply the function to the EdLevel column
dataset['EdLevel'] = dataset['EdLevel'].apply(clean_education)
```

```python
dataset['EdLevel'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_education = LabelEncoder()
dataset['EdLevel'] = le_education.fit_transform(dataset['EdLevel'])
dataset['EdLevel'].unique()
```

```python
le_country = LabelEncoder()
dataset['Country'] = le_country.fit_transform(dataset['Country'])
dataset['Country'].unique()
```

```python
X = dataset.drop("Salary", axis=1)
y = dataset["Salary"]
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(X, y.values)
```

```python
y_pred = linear_reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
error
```

```python
from sklearn.tree import DecisionTreeRegressor
dec_tree_reg = DecisionTreeRegressor(random_state=0)
dec_tree_reg.fit(X, y.values)
```

```python
y_pred = dec_tree_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg = RandomForestRegressor(random_state=0)
random_forest_reg.fit(X, y.values)
```

```python
y_pred = random_forest_reg.predict(X)
```

```python
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
from sklearn.model_selection import GridSearchCV

max_depth = [None,2,4,6,8,10,12]
parameters = {"max_depth":max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(X, y.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```
